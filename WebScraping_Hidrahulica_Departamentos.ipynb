## BLOQUE DE DESCARGA:
Con este bloque se extran la informacion de las tablas que corresponde a las precipitaciones de las distintas localidades de Entre-Rios pagina: https://www.hidraulica.gob.ar/redPluviometricaHist.php.
- La extraccion es por tabla,introduciendo en el codigo los imputs: fecha_inicio, fecha_fin y nombres. Estos resaltan en el codigo  mendiante comentarios, para su localizacion.
- Cada bloque corresponde a un departamento, y los datos que recolectan cada uno de ellos, son volcados al Dataframe "df0".
- Como ultima etapa del proceso, se descarga un archivo .csv


## Colón
En la extraccion de datos de la tabla de colon, tuvimos a que hacer una particularidades (ingenieria de datos), con respecto al resto de las tablas,ya que la exrtaccion no respetaba lo que necesitabamos.

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
from datetime import datetime, timedelta

# Función para generar una lista de fechas entre dos fechas dadas
def generar_lista_fechas(fecha_inicio, fecha_fin):
    fechas = []
    fecha_actual = fecha_inicio
    while fecha_actual <= fecha_fin:
        fechas.append(fecha_actual.strftime('%Y-%m-%d'))
        fecha_actual += timedelta(days=1)
    return fechas

# Crear DataFrame base para volcar todos los DataFrames generados en el ciclo
columnas = ['fecha', 'Estación', 'Precipitación']
df0 = pd.DataFrame(columns=columnas)

# Link de la página con la tabla
url = "https://www.hidraulica.gob.ar/redPluviometricaHist.php"

# Generar lista de fechas desde 2010-09-02 hasta 2024-05-22
fecha_inicio = datetime.strptime('2010-09-02', '%Y-%m-%d')
fecha_fin = datetime.strptime('2024-05-22', '%Y-%m-%d')
lista_fechas = generar_lista_fechas(fecha_inicio, fecha_fin)

# Iterar sobre cada fecha en la lista de fechas
for fecha in lista_fechas:
    # Parámetros del formulario para la fecha actual
    form_data = {
        'selectedFecha': fecha
    }

    # Realizar la solicitud POST para seleccionar la fecha
    response = requests.post(url, data=form_data)
    sopa = bs(response.content, 'html.parser')

    # Inicializar listas para guardar los datos de la fecha actual
    estaciones = []
    precipitaciones = []
    fechas = []

    # Encontrar todas las tablas en la página
    tablas = sopa.find_all('table')

    for tabla in tablas:
        # Verificar si la tabla corresponde a "Colón"
        encabezado = tabla.find('h3')
        if encabezado and encabezado.get_text(strip=True) == "Colón":
            # Encontrar todas las filas de la tabla
            filas = tabla.find_all('tr')
            for fila in filas[3:]:  # Empezar desde la cuarta fila (índice 3)
                # Encontrar todas las celdas de la fila
                celdas = fila.find_all('td')
                if len(celdas) > 1:  # Asegurarse de que la fila contiene datos útiles
                    estacion = celdas[0].get_text(strip=True)
                    precipitacion = celdas[1].get_text(strip=True)
                    # Agregar datos a las listas
                    estaciones.append(estacion)
                    precipitaciones.append(precipitacion)
                    fechas.append(fecha)  # Añadir la fecha actual

            # Crear un DataFrame con los datos obtenidos para esta fecha
            data = {'fecha': fechas, 'Estación': estaciones, 'Precipitación': precipitaciones}
            df = pd.DataFrame(data)

            # Filtrar solo las filas necesarias (las filas 4 a 7 del DataFrame)
            df_filtered = df.iloc[4:10]  # Ajustar índice para incluir solo las filas 4 a 7

            # Concatenar el DataFrame filtrado a df0 (original)
            df0 = pd.concat([df0, df_filtered], ignore_index=True)

#Eliminamos datos inencesarios
valores_a_eliminar = ["Dpto:", "Estación", "Dpto:Concordia", "Dpto:Diamante"]
df0 = df0[~df0["Estación"].isin(valores_a_eliminar)]
# Mostrar el DataFrame consolidado
print(df0)


# Guardar df0 en un archivo CSV
df0.to_csv('Colón.csv', index=False)
```
